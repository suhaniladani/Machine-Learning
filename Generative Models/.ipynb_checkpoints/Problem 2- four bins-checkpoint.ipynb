```python
def train_test_split(df, test_size):
    
    if isinstance(test_size, float):
        test_size = round(test_size * len(df))

    index_list = df.index.tolist()
    test_indexes = random.sample(population=index_list, k=test_size)

    test_df = df.loc[test_indexes]
    train_df = df.drop(test_indexes)
    
    return train_df, test_df
```

```python
# labels = ['min_value', 'low-mean-value', 'overall-mean-value', 'high-mean-value', 'max-value']
labels = ['1', '2', '3', '4']
```

```python
for j in df.columns[:-1]:
    mean = df[j].mean()
    mean1_df = df.loc[df['label'] == 1][j].mean()
    mean2_df = df.loc[df['label'] == 0][j].mean()
    if(mean1_df < mean2_df):
        lo = mean1_df 
        hi = mean2_df
    else:
        lo = mean2_df
        hi = mean1_df
    bins = [-1*np.inf, lo, mean, hi, np.inf]
    df[j] = df[j].replace(0,mean) #Replace 0 with mean
    df[j] = pd.cut(df[j],bins,labels=labels)
```

```python
def count(data,colname,label,target):
    condition = (data[colname] == label) & (data['label'] == target)
    return len(data[condition])
```

```python
probabilities = {0:{},1:{}}
```

```python
random.seed(0)
train, test = train_test_split(df, 0.20)

X_train = train
X_test = test.iloc[:,:-1]
y_test = test.iloc[:,-1]
```

```python
cnt_label_0 = count(X_train,'label',0,0)
cnt_label_1 = count(X_train,'label',1,1)
    
prob_0 = cnt_label_0/len(X_train)
prob_1 = cnt_label_1/len(X_train)
```

```python
cnt_label_1
```

```python
for col in X_train.columns[:-1]:
        probabilities[0][col] = {}
        probabilities[1][col] = {}
        
        for category in labels:
            count_ct_0 = count(X_train,col,category,0)
            count_ct_1 = count(X_train,col,category,1)
            
            probabilities[0][col][category] = (count_ct_0 +1) / (cnt_label_0+4)
            probabilities[1][col][category] = (count_ct_1+1) / (cnt_label_1+4)
```

```python
probabilities
```

```python
predicted = []
```

```python
for row in range(0,len(X_test)):
        prod_0 = prob_0
        prod_1 = prob_1
        for feature in X_test.columns:   #p of x give y
            prod_0 *= probabilities[0][feature][X_test[feature].iloc[row]]
            prod_1 *= probabilities[1][feature][X_test[feature].iloc[row]]
        
        #Predict the outcome
        if prod_0 > prod_1:
            predicted.append(0)
        else:
            predicted.append(1)
```

```python
tp,tn,fp,fn = 0,0,0,0
for j in range(0,len(predicted)):
    if predicted[j] == 0:
        if y_test.iloc[j] == 0:
            tp += 1
        else:
            fp += 1
    else:
        if y_test.iloc[j] == 1:
            tn += 1
        else:
            fn += 1
```

```python
accuracy = (tp + tn)/len(y_test)
```

```python
print('Accuracy:',((tp+tn)/len(y_test))*100)
```

```python
thresholds = np.linspace(2,-2,105)

ROC = np.zeros((105,2))

for i in range(105):
    t = thresholds[i]
    
    TP_t = np.logical_and( predicted > t, y_test==1 ).sum()
    TN_t = np.logical_and( predicted <=t, y_test==0 ).sum()
    FP_t = np.logical_and( predicted > t, y_test==0 ).sum()
    FN_t = np.logical_and( predicted <=t, y_test==1 ).sum()

    FPR_t = FP_t / float(FP_t + TN_t)
    ROC[i,0] = FPR_t

    TPR_t = TP_t / float(TP_t + FN_t)
    ROC[i,1] = TPR_t

fig = plt.figure(figsize=(6,6))
plt.plot(ROC[:,0], ROC[:,1], lw=2)
plt.xlabel('$FPR(t)$')
plt.ylabel('$TPR(t)$')
plt.grid()
```

```python
AUC = 0.
for i in range(100):
    AUC += (ROC[i+1,0]-ROC[i,0]) * (ROC[i+1,1]+ROC[i,1])
AUC *= 0.5
AUC
```
