```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2 #pip install opencv-python

#deep dreams of neural network
```

```python
DataDir = "V:\SEM4\Big_Data_Intelligence_Grading\workshop\PetImages"
Categories = ['Dog','Cat']
```

```python
for Category in Categories:
    path = os.path.join(DataDir, Category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap='gray')
        plt.show()
        break
```

```python
print(img_array)
```

```python
Img_Size = 50

new_array = cv2.resize(img_array, (Img_Size, Img_Size))
plt.imshow(new_array, cmap='gray')
plt.show()
```

```python
training_data = []


def create_training_data():
    for Category in Categories:
        path = os.path.join(DataDir, Category) #path to cats or dogs Category
        class_num = Categories.index(Category) # 0 = Dog, 1 = Cat
        for img in os.listdir(path):
            try:    
                img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (Img_Size, Img_Size))
                training_data.append([new_array, class_num])
            except Exception as e:
                pass
    
create_training_data()
```

```python
print(len(training_data))
```

```python
import random

random.shuffle(training_data)
```

```python
for sample in training_data[0:10]:
    print(sample[1])
```

```python
x = [] #feature
y = [] #label
```

```python
for features, label in training_data:
    x.append(features)
    y.append(label)
    
x = np.array(x).reshape(-1, Img_Size, Img_Size, 1) # -1 how many shapes we have
                                                   # To make it numpy array
                                                   # 1 is because it's a gray scale
```

```python
import pickle

pickle_out  = open("x.pickle", 'wb')
pickle.dump(x, pickle_out)
pickle_out.close()

pickle_out  = open("y.pickle", 'wb')
pickle.dump(y, pickle_out)
pickle_out.close()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import pickle
```

```python
pickle_in = open("X.pickle","rb")
X = pickle.load(pickle_in)

pickle_in = open("y.pickle","rb")
y = pickle.load(pickle_in)

X = X/255.0

model = Sequential()

model.add(Conv2D(64, (3, 3), input_shape=X.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors

# model.add(Dense(64))
# model.add(Activation('relu'))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X, y, batch_size=30, epochs=1, validation_split=0.3)
```
